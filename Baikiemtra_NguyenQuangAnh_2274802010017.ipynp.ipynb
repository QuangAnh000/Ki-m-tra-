```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Product': ['Laptop', 'Phone', 'Chair', 'Desk', 'Headphone', 'Monitor', 'Sofa', 'Keyboard', 'Mouse', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Accessories', 'Electronics', 'Furniture', 'Accessories', 'Accessories', 'Electronics'],
    'Price': [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-15', '2024-03-18', '2024-04-12', '2024-05-20', '2024-06-20', '2024-02-08', '2024-07-14', '2024-09-30'],
    'Quantity': [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}
df = pd.DataFrame(data)
df['Sales_Level'] = pd.cut(df['Price'], bins=[-float('inf'), 500, 2000, float('inf')], labels=['Low', 'Medium', 'High'])
sales_level_counts = df.groupby('Sales_Level')['Quantity'].sum()
plt.figure(figsize=(8, 5))
sales_level_counts.plot(kind='bar', color=['red', 'orange', 'green'])
plt.title('Number of Orders at Each Sales Level')
plt.xlabel('Sales Level')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.show()
```

```python
15. Tạo một cột mới <Discount> với giá trị giảm giá 10% nếu doanh thu (Revenue) > 1000, ngược lại giảm giá 5%.
```

```python
import pandas as pd
data = {
    'Product': ['Laptop', 'Phone', 'Chair', 'Desk', 'Headphone', 'Monitor', 'Sofa', 'Keyboard', 'Mouse', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Accessories', 'Electronics', 'Furniture', 'Accessories', 'Accessories', 'Electronics'],
    'Price': [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-15', '2024-03-18', '2024-04-12', '2024-05-20', '2024-06-20', '2024-02-08', '2024-07-14', '2024-09-30'],
    'Quantity': [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}
df = pd.DataFrame(data)
df['Revenue'] = df['Price'] * df['Quantity']
df['Discount'] = df['Revenue'].apply(lambda x: 0.10 if x > 1000 else 0.05)
print(df)
```

```python
16. Tính doanh thu sau giảm giá <Final_Revenue> bằng công thức:

Final_Revenue = Revenue - ( Revenue ×Discount ).

Tạo một cột mới trong Dataframe với tên <Final_Revenue> gồm doanh thu sau giảm giá.
```

```python
import pandas as pd
data = {
    'Product': ['Laptop', 'Phone', 'Chair', 'Desk', 'Headphone', 'Monitor', 'Sofa', 'Keyboard', 'Mouse', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Accessories', 'Electronics', 'Furniture', 'Accessories', 'Accessories', 'Electronics'],
    'Price': [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-15', '2024-03-18', '2024-04-12', '2024-05-20', '2024-06-20', '2024-02-08', '2024-07-14', '2024-09-30'],
    'Quantity': [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}
df = pd.DataFrame(data)
df['Revenue'] = df['Price'] * df['Quantity']
df['Discount'] = df['Revenue'].apply(lambda x: 0.10 if x > 1000 else 0.05)
df['Final_Revenue'] = df['Revenue'] - (df['Revenue'] * df['Discount'])
print(df)
```

```python
import pandas as pd
data = {
    'Product': ['Laptop', 'Phone', 'Chair', 'Desk', 'Headphone', 'Monitor', 'Sofa', 'Keyboard', 'Mouse', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Accessories', 'Electronics', 'Furniture', 'Accessories', 'Accessories', 'Electronics'],
    'Price': [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-15', '2024-03-18', '2024-04-12', '2024-05-20', '2024-06-20', '2024-02-08', '2024-07-14', '2024-09-30'],
    'Quantity': [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Revenue'] = df['Price'] * df['Quantity']
df['Discount'] = df['Revenue'].apply(lambda x: 0.10 if x > 1000 else 0.05)
df['Final_Revenue'] = df['Revenue'] - (df['Revenue'] * df['Discount'])
df.to_csv('BKT_22748020100170_NguyenQuangAnh.csv', index=False)
print("DataFrame has been saved to 'BKT_22748020100170_NguyenQuangAnh.csv'.")
```
